```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# membuat dataframe
data = {
    'Kabupaten': ['Lampung Barat', 'Tanggamus', 'Lampung Selatan', 'Lampung Timur', 'Lampung Tengah', 'Lampung Utara', 'Way Kanan','Tulang Bawang','Pesawaran','Pringsewu','Mesuji','Tulang Bawang Barat','Pesisir Barat','Bandar Lampung','Metro'],
    'Jumlah Sekolah Tk (Negeri)': [18, 3, 5, 1, 8, 31, 2, 3, 8, 5, 4, 1, 12, 4, 4,]
}

df = pd.DataFrame(data)
```

```python
# Membuat bar chart dengan Matplotlib
plt.figure(figsize=(12, 6))
plt.bar(df['Kabupaten'], df['Jumlah Sekolah Tk (Negeri)'], color='skyblue')
plt.title('Jumlah Sekolah Taman Kanak-Kanak per Kabupaten')
plt.xlabel('Kabupaten')
plt.ylabel('Jumlah Sekolah Tk (Negeri)')
plt.xticks(rotation=45)
plt.show()
```

**Tujuan**
--
* Mudah Dibaca: Bar chart memiliki representasi visual yang jelas, sehingga memudahkan para pembaca untuk memahami perbedaan antara kategori dengan mudah.
* Efisien : Bar chart mudah dipahami dan oleh karena itu memudahkan untuk mnyampaikan informasi dengan sederhana.

```python
# Membuat plot dengan seaborn
plt.figure(figsize=(12,6))
sns.lineplot(x='Kabupaten', y='Jumlah Sekolah Tk (Negeri)', data=df, marker='o')
plt.title('Jumlah Sekolah Taman Kanak-Kanak per Kabupaten')
plt.xlabel('Kabupaten')
plt.ylabel('Jumlah Sekolah Tk (Negeri)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

**Tujuan**
--
* Efisien : Memudahkan membuat plot kompleks dengan sintaks yang sederhana sehingga, user dapat dengan mudah dan cepat menghasilkan visualisasi yang informatif.
* Estetika : Secara visualisasi menarik, sehingga dapat mempresntasikan data dengan baik.
